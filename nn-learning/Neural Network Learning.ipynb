```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt

from scipy.io import loadmat

from display_data import *
from rand_mat import *
from sigmoid import *
from nn_cross_entropy import *
```

```python
data = loadmat('ex4data.mat')
X = data['X']
y = data['y']

m, n = X.shape
input_layer_size = 400
hidden_layer_size = 25
labels = 10

random_indices = np.random.randint(0, m, size=100)
vis_data = X[random_indices, :]
```

```python
display_data(vis_data);
```

```python
theta1 = rand_mat(hidden_layer_size, input_layer_size + 1)
theta2 = rand_mat(labels, hidden_layer_size + 1)
nn_params = np.concatenate((theta1.flatten(), theta2.flatten()))
Lambda = 0
```

```python
nn_cross_entropy(nn_params, input_layer_size, hidden_layer_size, labels, X, y, Lambda)
```

```python
options = {'maxiter':200}
args = (input_layer_size, hidden_layer_size, labels, X, y, Lambda)
opt.minimize(fun=nn_cross_entropy, x0=nn_params, method='TNC', jac=nn_cross_entropy_grad, options=options, args=args)
```
